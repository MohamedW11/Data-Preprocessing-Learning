## Importing

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

plt.style.use('default')
```

## Some key concepts
##### Figure : the container - entire page
##### Axes : the actual plot area
##### Axis : the number line objects (x-axis, y-axis)

## 1) Create very simple plot

```python
x = np.random.randint(0,20,2)
y = np.random.randint(0,20,2)

plt.plot(x,y)
plt.plot(y,x)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Simple Line Plot")

#plt.show() not needed because the first cell
```

## 2) Subplots
##### the idea of the subplots basically we have the figure which is the whole page, while the axes is the one plotted inside it, so the subplot is about many axes in the same figure

```python
fig, ax = plt.subplots(1 , 2) # take the rows and the columns 
ax[0].plot([1,2,3,4,5] , [5,9,5,6,2])
ax[1].plot([5,6,8,4,6] , [5,6,7,5,6])
```

## 3) Basic line Plots

```python
x = np.linspace(0,10,100) # 100 points from 0 - 10
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize=(8, 5)) # width and height
plt.plot(x,y1 , label = "sin(x)")
plt.plot(x,y2 , label = "cos(x)")
plt.plot()
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Sin, Cos Waves")
plt.legend() # show the part in the above right
plt.grid(True)
```

## 4) Customizing Plots

```python
x = np.linspace(0, 10, 20)

plt.figure(figsize=(12, 6))

plt.plot(x, x, 'b-', label='Solid line', linewidth=2)
plt.plot(x, x+1, 'r--', label='Dashed line', linewidth=2)
plt.plot(x, x+2, 'g-.', label='Dash-dot line', linewidth=2)
plt.plot(x, x+3, 'm:', label='Dotted line', linewidth=2)
plt.plot(x, x+4, 'co-', label='Circles', markersize=8)
plt.plot(x, x+5, 's-', label='Squares', markersize=6)
plt.plot(x, x+6, '^-', label='Triangles', markersize=6)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Styles and Markers')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Common format string: '[color][marker][line]'
# Colors: b (blue), r (red), g (green), c (cyan), m (magenta), y (yellow), k (black)
# Markers: o (circle), s (square), ^ (triangle), * (star), + (plus), x (x)
# Lines: - (solid), -- (dashed), -. (dash-dot), : (dotted)
```

## 5) Scatter Plot

```python
np.random.seed(42)
x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.5

plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.6)  # alpha controls transparency
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter Plot')
plt.grid(True, alpha=0.3)
```

```python
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)  # Random colors
sizes = 1000 * np.random.rand(100)  # Random sizes


plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='Set1')
plt.colorbar(label='Color intensity')  # Add colorbar
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter Plot with Color and Size Variations')
plt.grid(True, alpha=0.3)
plt.show()
```

## 6) Bar Chart

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]

plt.figure(figsize=(8, 6))
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Chart')
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

```python
# Horizontal bar chart
plt.figure(figsize=(8, 6))
plt.barh(categories, values, color='coral', edgecolor='black', linewidth=1.2)
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart')
plt.grid(True, alpha=0.3, axis='x')
```

```python
categories = ['Q1', 'Q2', 'Q3', 'Q4']
product_a = [20, 35, 30, 35]
product_b = [25, 32, 34, 20]
product_c = [30, 25, 30, 25]

x = np.arange(len(categories))
width = 0.25

plt.figure(figsize=(10, 6))
plt.bar(x - width, product_a, width, label='Product A', color='#FF6B6B')
plt.bar(x, product_b, width, label='Product B', color='#4ECDC4')
plt.bar(x + width, product_c, width, label='Product C', color='#95E1D3')

plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.title('Quarterly Sales Comparison')
plt.xticks(x, categories)
plt.legend()
plt.grid(True, alpha=0.3, axis='y')
```
