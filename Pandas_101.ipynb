# 4. Missing Values

```python
# count missing values in every column
df.isnull().sum()
```

```python
# fill missing values with zero
df.fillna(0)
```

```python
# fill missing values with mean
df["age"].fillna(df["age"].mean())
```

```python
# drop rows with missing values
df.dropna(inplace=True)
```

# 5. Update

```python
# create new column
df["passed"] = True
```

```python
# update existing column
df["gpa"] = df["gpa"] + 0.5
```

```python
# update rows with condition
df.loc[df["gpa"] > 3.5, "status"] = "Excellent"
```

```python
# rename columns
df.rename(columns={"gpa":"GPA"})
```

```python
# change data type
df["age"] = df["age"].astype(float)
```

# 6. Delete

```python
# drop column
df.drop("projects", axis=1)
```

```python
# drop rows
df.drop([0,1,2])
```

```python
# remove duplicated rows
df.drop_duplicates()
```

# 9. GroupBy

```python
# get average GPA per country
df.groupby("country")["gpa"].mean()
```
