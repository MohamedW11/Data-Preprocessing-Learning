```python
import pandas as pd
```

### 4) work with every column, see which one need cleaning
#### for example the last name has many errors need to be clearn

```python
df["Last_Name"]  
```

```python
# this line make us replace any characters we need by empty space
df["Last_Name"] = df["Last_Name"].str.replace(r"[._/]", "", regex=True)
df
```

#### then work with the phone numbers column

```python
df["Phone_Number"]
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace(r"\D", "", regex=True)
```

```python
df["Phone_Number"] = df["Phone_Number"].fillna(0)
```

```python
df["Phone_Number"] = df["Phone_Number"].replace("" , 0)
```

```python
df["Phone_Number"]
```

```python
formatted_numbers = []

for phone in df["Phone_Number"]:
    phone = str(phone)      # Make sure it's a string

    formatted = ""

    for i, digit in enumerate(phone):
        formatted += digit

        if i == 2 or i == 5:
            formatted += "-"

    formatted_numbers.append(formatted)

df["Phone_Number"] = formatted_numbers
```

```python
df
```

#### handle the address column by seperating

```python
df["Address"] = df["Address"].replace("N/a", pd.NA)
df["Address"] = df["Address"].fillna("Unknown")
df["Address"] = df["Address"].str.title()


df["House_Number"] = df["Address"].str.extract(r"^(\d+)")
df["ZIP_Code"] = df["Address"].str.extract(r"(\d{5})$")
df["City"] = df["Address"].str.extract(r",\s*([^,]+?)(?:,\s*\d{5})?$")
df["Street"] = (
    df["Address"]
    .str.replace(r"^\d+\s*", "", regex=True)
    .str.replace(r",.*", "", regex=True)
)
df["Street"] = df["Street"].str.replace(r"\s+", " ", regex=True).str.strip()
```

```python
df
```

#### Paying Customer column

```python
df["Paying Customer"]
```

```python
df["Paying Customer"] = df["Paying Customer"].replace(["Unknown", "Y"], "Yes")
df["Paying Customer"] = df["Paying Customer"].replace("N", "No")
df["Paying Customer"]
```

#### Do_Not_Contact Column

```python
df["Do_Not_Contact"]
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace("Y" , "Yes")
df["Do_Not_Contact"]
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace("N" , "No")
df["Do_Not_Contact"]
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].fillna("Yes")
df["Do_Not_Contact"]
```
